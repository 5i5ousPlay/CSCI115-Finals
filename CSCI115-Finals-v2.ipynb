```python
def plot_belief_over_time(beliefs_over_time, model, carrying_capacity, 
                          belief_threshold=0.5, disbelief_threshold=-0.5,
                          title="Belief Over Time",
                         show_pop=True):
    believers_over_time = []
    nonbelievers_over_time = []
    moderates_over_time = []
    if show_pop:
        pop_over_time = [x.shape[0] for x in beliefs_over_time]
    
    for timestep in beliefs_over_time:
        believers = (timestep > belief_threshold).sum(dim=0)
        believers_over_time.append(believers)
    
        nonbelievers = (timestep < disbelief_threshold).sum(dim=0)
        nonbelievers_over_time.append(nonbelievers)
    
        total_pop = timestep.shape[0]
        moderates = total_pop - (nonbelievers + believers)
        moderates_over_time.append(moderates if moderates > 0 else torch.tensor([0]))
    
    fig, ax = plt.subplots(figsize=(10,8))
    ax.plot(range(0,model.num_timesteps+1), believers_over_time, color='blue', label='Believers')
    ax.plot(range(0,model.num_timesteps+1), nonbelievers_over_time, color='red', label='Non-believers')
    ax.plot(range(0,model.num_timesteps+1), moderates_over_time, color='green', label='Unsure')
    if show_pop:
        ax.plot(range(0,model.num_timesteps+1), pop_over_time, linestyle='dashdot',
               color='gold', label="Population Size")
        ax.hlines(carrying_capacity, 0, model.num_timesteps+1, linestyle='dashed', color='orange',
                 label='Carrying Capacity')
    ax.legend(title='Groups', loc='center right')
    ax.set_title(title)
    ax.set_xlabel("Timestep")
    ax.set_ylabel("Count")
    ax.grid(True)
```

```python
plot_belief_over_time(over_time, model, 10000)
```

```python
size = len(G.nodes)
PC = (size * (size-1))/2
density_ot = np.array([x.shape[1]/PC for x in edge_over_time])
density_growth = np.diff(density_ot)
window_size = 5
moving_avg_density_ot = np.convolve(density_ot, np.ones(window_size)/window_size, mode='valid')
fig, ax = plt.subplots()
ax.plot(range(0, len(over_time)), density_ot, label='Density')
ax.plot(range(1, len(density_growth) + 1), density_growth, label='Density Growth')
ax.plot(range(window_size - 1, len(density_ot)), moving_avg_density_ot, label="Moving Average", color='red', linestyle='dashed')
ax.set_title("Density over Time")
ax.set_xlabel("Timestep")
ax.set_ylabel("Density")
ax.legend()
ax.grid(True)
```

```python
def get_communities_modularity_over_time(edgelist_over_time):
    communities_over_time = []
    modularity_over_time = []
    for edgelist in tqdm(edgelist_over_time, desc="Processing Edgelists"):
        el = list(map(tuple, edgelist.T.numpy()))
        G = nx.from_edgelist(el)
        communities = nx.community.louvain_communities(G, seed=42)
        communities_over_time.append(communities)
        modularity_over_time.append(nx.community.modularity(G, communities))
    return communities_over_time, modularity_over_time
```

```python
communities_over_time, modularity_over_time = get_communities_modularity_over_time(edge_over_time)
```

```python
fig, ax = plt.subplots(nrows=2,figsize=(10,12))
community_count = [len(communities) for communities in communities_over_time]
window_size = 5
moving_avg_community_count = np.convolve(community_count, np.ones(window_size)/window_size, mode='valid')
ax[0].plot(range(0, len(over_time)), community_count, label='Community Count')
ax[0].plot(range(window_size - 1, len(community_count)), moving_avg_community_count, 
        label='Community Count (Moving Avg.)', color='red', linestyle='dashed')
ax[0].set_title("Community Count Over Time")
ax[0].set_xlabel("Timestep")
ax[0].set_ylabel("Community Count")
ax[0].legend()
ax[0].grid(True)

avg_community_size = []
for communities in communities_over_time:
    community_sizes = np.array([len(community) for community in communities])
    avg_community_size.append(community_sizes.mean())
moving_avg_community_size = np.convolve(avg_community_size, np.ones(window_size)/window_size,
                                        mode='valid')
ax[1].plot(range(0, len(over_time)), avg_community_size, label='Community Size')
ax[1].plot(range(window_size - 1, len(avg_community_size)), moving_avg_community_size,
          label='Community Size (Moving Avg.)', color='red', linestyle='dashed')
ax[1].set_title("Average Community Size")
ax[1].set_xlabel("Timestep")
ax[1].set_ylabel("Avg. Community Size")
ax[1].grid(True)
ax[1].legend()
```

```python
def node_history_to_df(node_histories):
    data = []
    for node in node_histories:
        history = np.array(node_histories[node])
        # if history.ndim < 2 or history.shape[0] == 0:
        #     continue
            
        openness = history[0,1]
        c_bound = history[0,2]
        
        # belief
        belief = history[:, 0]
        avg_belief = belief.mean()
        std_belief = belief.std()
        start_belief = belief[0]
        last_belief = belief[-1]
        range_belief = np.ptp(belief)
        belief_mean_step_grad = np.diff(belief).mean() if len(belief) > 1 else 0.0
        
        # degree
        deg = history[:, 3]
        avg_deg = deg.mean()
        std_deg = deg.std()
        start_deg = deg[0]
        last_deg = deg[-1]
        range_deg = np.ptp(deg)
        deg_mean_step_grad = np.diff(deg).mean() if len(deg) > 1 else 0.0
    
        # lifetime
        lifetime = history.shape[0]

        row = np.array([avg_belief, std_belief, start_belief, last_belief, range_belief, belief_mean_step_grad,
               avg_deg, std_deg, start_deg, last_deg, range_deg, deg_mean_step_grad, 
               openness, c_bound, lifetime])
        data.append(row)
        # print(data)

    columns = ['avg_belief', 'std_belief', 'start_belief', 'last_belief', 'range_belief', 'belief_mean_step_grad',
              'avg_deg', 'std_deg', 'start_deg', 'last_deg', 'range_deg', 'deg_mean_step_grad', 'openness', 'c_bound', 'lifetime']
    df = pd.DataFrame(data, columns=columns)
    return df
```

```python
df = node_history_to_df(node_histories)
```

```python
from sklearn.cluster import DBSCAN, OPTICS, KMeans
from sklearn.metrics import silhouette_score
from collections import Counter
from sklearn.preprocessing import StandardScaler
```

```python
min_samples = range(10,21)
eps = np.arange(0.05,0.13, 0.01)

output = []
sc = StandardScaler()
stdDf = pd.DataFrame(sc.fit_transform(df), columns=df.columns)
min_samples = range(10,20)
for ms in tqdm(min_samples, desc="Testing Min Samples"):
    labels = OPTICS(min_samples=ms).fit(stdDf).labels_
    score = silhouette_score(stdDf, labels)
    output.append((ms, score))
```

```python
min_samples, score = sorted(output, key=lambda x:x[-1])[-1]
print(f"Best silhouette_score: {score}")
print(f"min_samples: {min_samples}")
```

```python
labels = OPTICS(min_samples=min_samples).fit(stdDf).labels_
clusters = len(Counter(labels))
print(f"Number of clusters: {clusters}")
print(f"Number of outliers: {Counter(labels)[-1]}")
print(f"Silhouette_score: {silhouette_score(stdDf, labels)}")
```

```python
df['optics_label'] = labels
```

```python
corr = df.select_dtypes('number').corr()
fig, ax = plt.subplots(figsize=(15, 10))
sns.heatmap(corr, annot=True, ax=ax)
```

```python
inertia = []
for i in range(2, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(stdDf)
    inertia.append(kmeans.inertia_)

fig, ax = plt.subplots()
ax.plot(range(2,11), inertia, marker='o')
ax.set_title("Inertia")
```

```python
silhouette_scores = []
for i in tqdm(range(2, 11), desc='Testing KMeans Silhoette Score'):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(stdDf)
    ss = silhouette_score(stdDf, kmeans.labels_)
    silhouette_scores.append(ss)
```

```python
fig, ax = plt.subplots()
ax.plot(range(2,11), silhouette_scores, marker='o')
ax.set_title("Silhouette Scores")
```
